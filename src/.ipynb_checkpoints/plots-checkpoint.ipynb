```python
df = session.read.parquet("/bda/anx_adv_achurn/t_bda_anx_dt_ach_b2b_master_table_train_20200918").cache()
```

```python
df_sottocanali_num_churn = df.groupBy('mds_sottocanale', 'val_churn_2_months').count().where('val_churn_2_months = 1.0').select('mds_sottocanale', 'count').toPandas()
```

```python
dict_of_churn_counts = df_sottocanali_num_churn.set_index('mds_sottocanale')['count'].to_dict()
dict_of_churn_counts
```

```python
with open(os.getcwd()+"/../Sviluppo_US/ITEAC-2154_Evolutiva_Modello_predittivo_anti-churn_B2B/models/gensim_word2vec_model_small_medium.pickle", 'rb') as file:
    gensim_word2vec_model = pickle.load(file)
```

```python
dict_of_embeddings_dicts = {}
for column in tqdm(['mds_sottocanale']):
    
    # Get all the distinct values of the column
    distinct_values = list(df.select(column).distinct().toPandas()[column].values)
    
    # Create a dictionary of embeddings
    embeddings = {}
    for value in distinct_values:
        embeddings[value] = gensim_word2vec_model[value]
    
    dict_of_embeddings_dicts[column] = embeddings
```

```python
len(dict_of_churn_counts)
```

```python
for item in dict_of_embeddings_dicts['mds_sottocanale'].keys():
    if item not in dict_of_churn_counts:
        dict_of_churn_counts[item] = 0
```

```python
len(dict_of_churn_counts)
```

```python
def plot_embeddings_TSNE(w, h, embeddings_dict, churn_count_dict, size, label, cmap, title):
    
    arr = np.empty((0,size), dtype='f')
    word_labels = []
    color = []
    
    for wrd in embeddings_dict.keys():
        vector = embeddings_dict[wrd]
        word_labels.append(wrd)
        color.append(churn_count_dict[wrd])
        arr = np.append(arr, np.array([vector]), axis=0)
        
    tsne = TSNE(n_components=2, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arr)
    
    #figure
    fig, ax1 = plt.subplots()
    fig.set_size_inches(w, h)
    
    #labels
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title(title)
    
    #c sequence
    c = color
    
    #plot
    x_coords = Y[:, 0]
    y_coords = Y[:, 1]
    plt.scatter(x_coords, y_coords, c=c, cmap = cmap, s = [(item+2)*.5 for item in color], alpha =0.5)
    cbar = plt.colorbar()
    cbar.set_label('Number of churn')

    if label:
        for label, x, y in zip(word_labels, x_coords, y_coords):
            plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
            
    plt.show()
```

```python
plot_embeddings_TSNE(18, 15, dict_of_embeddings_dicts['mds_sottocanale'], dict_of_churn_counts, size = 6, label = False, cmap="RdPu_r", title = "Embeddings plot mds_sottocanale")#nipy_spectral
```
